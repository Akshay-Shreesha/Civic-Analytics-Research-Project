```python
cols=['NYC Borough, Block and Lot (BBL)','NYC Building Identification Number (BIN)','Primary Property Type - Portfolio Manager-Calculated',
     'Office - Gross Floor Area (ft²)','Office - Worker Density (Number per 1,000 sq ft)','Total GHG Emissions Intensity (kgCO2e/ft²)','Latitude','Longitude','Census Tract']
```

### Offices

```python
office=LL84[LL84['Office - Worker Density (Number per 1,000 sq ft)']!='Not Available']
```

```python
office.shape
```

```python
office=office[cols]
office.head()
```

```python
office['Total GHG Emissions Intensity (kgCO2e/ft²)']
```

```python
office=office[office['Total GHG Emissions Intensity (kgCO2e/ft²)']!='Not Available']
```

```python
office.info()
```

```python
office['Office - Worker Density (Number per 1,000 sq ft)'] = office['Office - Worker Density (Number per 1,000 sq ft)'].apply(pd.to_numeric,errors='coerce')
```

```python
office['Total GHG Emissions Intensity (kgCO2e/ft²)']=office['Total GHG Emissions Intensity (kgCO2e/ft²)'].apply(pd.to_numeric,errors='coerce')
```

```python
office['Office - Gross Floor Area (ft²)']=office['Office - Gross Floor Area (ft²)'].apply(pd.to_numeric,errors='coerce')
```

```python
office.info()
```

```python
plt.scatter(x='Office - Worker Density (Number per 1,000 sq ft)',y='Total GHG Emissions Intensity (kgCO2e/ft²)',data=office)
```

```python
office.head()
```

```python
office['Total Workers']=office['Office - Gross Floor Area (ft²)']*office['Office - Worker Density (Number per 1,000 sq ft)']
```

```python
office['Total Emissions']=office['Total GHG Emissions Intensity (kgCO2e/ft²)']*office['Office - Gross Floor Area (ft²)']
```

```python
office['Per capita emissions']=office['Total Emissions']/office['Total Workers']
```

```python
office.head()
```

```python
office=office.iloc[:100,:]
```

```python
x=office['Office - Worker Density (Number per 1,000 sq ft)']
y=office['Per capita emissions']

# Initialize layout
fig, ax = plt.subplots(figsize = (6, 3))

plt.scatter(x,y)

b, a = np.polyfit(x, y, deg=1)
xseq = np.linspace(0, 10, num=100)

plt.xlabel("Office - Worker Density (Number per 1,000 sq ft)")
plt.ylabel("Per capita emissions")
plt.title("Energy Efficiency")

# Plot regression line
ax.plot(xseq, a + b * xseq, color="k", lw=2)
#plt.savefig('Energy Efficiency.png', bbox_inches='tight')
```
